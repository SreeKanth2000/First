```python
link='https://www.flipkart.com/laptops/~best-15-inch-laptops/pr?sid=6bo%2Cb5g'
page=requests.get(link) #1 web request is sent to the targeted website to collect the required data
print(page)
```

```python
#create list
products = [] #list to store names of laptops
prices = [] #list to store laptop prices
ratings = [] #list to store ratings
for data in soup.find_all('a', class_ = '_1fQZEK'):
    name = data.find('div', attrs = {'class' : '_4rR01T'})
    price = data.find('div', attrs = {'class' : '_30jeq3 _1_WHN1'})
    rating = data.find('div', attrs = {'class' : '_3LWZlK'})
    
    products.append(name.text)
    prices.append(price.text)
    ratings.append(rating.text) if type(rating)==bs4.element.Tag else ratings.append('NaN')
```

```python
import pandas as pd
df = pd.DataFrame({'Product Name':products, 'Price':prices, 'Rating':ratings})

```
